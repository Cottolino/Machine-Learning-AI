```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
plt.rcParams["figure.figsize"] = [12,8]
```

```python
stock_price = np.arange(30,130) + np.random.rand(100)*10
```

```python
plt.title("Valore del titolo per i primi 100 giorni di trading")
plt.xlabel("Giorno")
plt.ylabel("Valore in Euro")
plt.grid()
#plt.xlim(40,50)
plt.annotate("Raggiounti 1000 utenti registrati", xy=(40,80), xytext=(10,90), arrowprops=dict(facecolor="black", shrink=0.05))
plt.text(84, 34, "Testo", size=15, bbox=dict(boxstyle="square,pad=0.3"))
plt.plot(stock_price)
```

```python
x = np.linspace(0,100, num=11)
y = np.power(x,2)
plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```

```python
x = np.arange(0.0, 5.0, 0.01)
plt.axes([0.,0.,0.5,1.0])
cos_x = np.cos(2*np.pi*x)
plt.plot(x,cos_x)

plt.axes([0.5,0.0,0.5,1.0])
sin_x = np.sin(2*np.pi*x)
plt.plot(x, sin_x)

plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,cos_x)
plt.subplot(1,2,2)
plt.plot(x, sin_x)
plt.tight_layout()
plt.show()
```
