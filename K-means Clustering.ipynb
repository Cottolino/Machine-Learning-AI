```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams["figure.figsize"] = (12,8)
```

```python
sns.set()
```

```python
from sklearn.datasets import make_blobs
```

```python
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=.8, random_state=0)
```

```python
plt.scatter(X[:,0], X[:,1], s=50)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(X)
```

```python
y = kmeans.predict(X)
```

```python
y[:10]
```

```python
centroids = kmeans.cluster_centers_
```

```python
plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap="viridis")
plt.scatter(centroids[:,0], centroids[:,1], c="red", s=200, alpha=0.5)
```

### K non conosciuto a priori

#### Creiamo 10 modelli di K

```python
ssd = {}
```

```python
for k in range(1,10):
    kmeans = KMeans(init="k-means++", n_clusters=k)
    kmeans.fit(X)
    ssd[k] = kmeans.inertia_
```

```python
plt.plot(list(ssd.keys()), list(ssd.values()), marker="o")
plt.xlabel("Numero di Clusters", fontsize=16)
plt.ylabel("Somma delle distanze al quadrato", fontsize=16)
plt.show()
```

## Per ottenere il valore di K utilizziamo il metodo Elbow Method
