```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mq = np.random.randint(110,size=50)+10
```

```python
noise = np.random.rand(50)*25000
```

```python
price = np.power(mq , 2) * 10
```

```python
price = price + noise
```

```python
plt.rcParams['figure.figsize'] = [12,8]
```

```python
mq_center = np.random.randint(110, size=10)+10
price_center = np.power(mq_center, 2)*15
noise_center= np.random.rand(10)*25000
price_center = price_center + noise_center
```

```python
plt.title("Rapporto Dimensione-Valore di monolocali")
plt.xlabel("Dimensione in metri quadrati")
plt.ylabel("Valore in Euro")
plt.grid()
plt.scatter(mq, price, label="monolocali in periferia", marker="s")
plt.scatter(mq_center, price_center, c="green", s=100., label="monolocali in centro", marker="*")
plt.legend()
plt.savefig("scatter.png")
```
