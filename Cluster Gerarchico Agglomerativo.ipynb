```python
import matplotlib.pyplot as plt
import seaborn as  sns
```

```python
plt.rcParams["figure.figsize"] = (12, 8)
```

```python
sns.set()
```

```python
from sklearn.datasets import make_blobs
```

```python
X, _ = make_blobs(n_samples=100, centers=3, cluster_std=0.5, random_state=0)
```

```python
plt.scatter(X[:,0], X[:,1], s=50)
```

### Costruzione Endrogramma

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
import pandas as pd
```

```python
link_matrix = linkage(X, method="ward")
```

```python
pd.DataFrame(link_matrix)
```

```python
dendrogram(link_matrix)
```

### Scegliamo una soglia per estrarre i cluster sottostanti

#### Scegliam una soglia di 10, cosÃ¬ avremo 3 cluster, perchÃ¨ taglia 3 linee

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=3)
```

```python
y = hc.fit_predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, s=200, cmap="viridis", edgecolors="black")
```
