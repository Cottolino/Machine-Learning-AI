```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
```

```python
from mnist import load_mnist
```

```python
X_train, X_test, Y_train, Y_test = load_mnist(path="MNIST")
```

```python
X_train.shape
```

```python
X_test.shape
```

## Input Rete Neurali sempre tra 0 e 1

### Normalizziamo

```python
mms = MinMaxScaler()
```

```python
X_train = mms.fit_transform(X_train)
```

```python
X_test = mms.transform(X_test)
```

## Verifichaimo se serva una rete

### Facciamo una regressione logistica semplice

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, Y_train)
```

### Calcolo metriche

```python
y_pred_train = lr.predict(X_train)
```

```python
y_prob_train = lr.predict_proba(X_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_prob = lr.predict_proba(X_test)
```

```python
accuracy_train = accuracy_score(Y_train, y_pred_train)
```

```python
accuracy_test = accuracy_score(Y_test, y_pred)
```

```python
loss_train = log_loss(Y_train, y_prob_train)
```

```python
loss_test = log_loss(Y_test, y_prob)
```

```python
print("ACCURACY TRAIN=%.4f TEST=%.4f" % (accuracy_train, accuracy_test))
print("LOG LOSS TRAIN=%.4f TEST=%.4f" % (loss_train, loss_test))
```

### Proviamo Rete Neurale

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(100,) , verbose=True)
```

```python
mlp.fit(X_train, Y_train)
```

```python
y_pred_train = mlp.predict(X_train)
y_prob_train = mlp.predict_proba(X_train)
y_pred = mlp.predict(X_test)
y_prob = mlp.predict_proba(X_test)
```

```python
accuracy_train = accuracy_score(Y_train, y_pred_train)
accuracy_test = accuracy_score(Y_test, y_pred)
loss_train = log_loss(Y_train, y_prob_train)
loss_test = log_loss(Y_test, y_prob)
```

```python
print("ACCURACY TRAIN=%.4f TEST=%.4f" % (accuracy_train, accuracy_test))
print("LOG LOSS TRAIN=%.4f TEST=%.4f" % (loss_train, loss_test))
```
