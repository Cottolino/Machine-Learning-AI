```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set()
```

```python
cols = ["label","alcol","acido malico","cenere","alcalinità della cenere","magnesio",
        "fenoli totali","flavonoidi","fenoli non-flavonoidi","proantocianidine", 
        "intensità del colore","tonalità", "OD280/OD315 dei vini diluiti","prolina"]
```

```python
wines = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", 
                 names=cols)
```

```python
wines.head()
```

```python
X = wines.drop("label", axis=1).values
```

```python
Y = wines["label"].values
```

### Standardizzazione

```python
ss = StandardScaler()
```

```python
X = ss.fit_transform(X)
```

### Pricipal Component Analysis [PCA]

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

### Fit => Calcola le direzioni principali che variano di piu [Nuovi Assi del Sistema di Riferimento]
### Transform => Proietta ogni nuovo punto sul nuovo Sistema Riferimento [Meno Dimensioni]

```python
X_pc = pca.fit_transform(X)
```

```python
X_pc
```

```python
plt.figure(figsize=(14,10))
plt.xlabel("Prima Componente Principale")
plt.ylabel("Seconda Componente Principale")
plt.scatter(X_pc[:,0], X_pc[:,1], c=Y, edgecolor="black")
```
