```python
from prophet import Prophet
```

```python
import mysql.connector
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
conn = mysql.connector.connect(
    host='localhost',
    port=3308,
    user='root',
    password='',
    database='mermec'
)
```

```python
query = """
SELECT datef AS date_invoice, total_ttc AS amount_total
FROM llx_facture
WHERE fk_statut IN (1, 2) -- 1 = Validata, 2 = Pagata
"""
```

```python
df = pd.read_sql(query, conn)
```

```python
conn.close()
```

```python
df.head(50)
```

```python
df['date_invoice'] = pd.to_datetime(df['date_invoice'])
```

```python
df.shape
```

```python
df = df.set_index('date_invoice').resample('MS').sum().reset_index()
```

```python
df.columns = ['ds', 'y']
```

```python
model = Prophet()
```

```python
model.fit(df)
```

```python
future = model.make_future_dataframe(periods=6, freq='MS')
```

```python
forecast = model.predict(future)
```

```python
model.plot(forecast)
plt.title("Previsione Fatturato Mensile da Dolibarr")
plt.xlabel("Data")
plt.ylabel("Fatturato")
plt.show()
```
